```python
def read_file(file):
    rules = []
    atualizations = []
    flag = False
    with open(file, 'r') as f:
        for line in f:
            if line == '\n':
                flag = True
                continue

            if not flag:
                rules.append(line.strip())
            else:
                atualizations.append(line.strip())
                
    return rules, atualizations

rules, atualizations = read_file('inputs/5.txt')
len(rules), len(atualizations)
```

```python
from collections import defaultdict, deque

# grafos para as dependÃªncias
pages_before = defaultdict(set)
pages_after = defaultdict(set)

for rule in rules:
    a, b = rule.split('|')
    a, b = int(a), int(b)
    
    pages_before[b].add(a)
    pages_after[a].add(b)
```

```python
for x in pages_after:
    print(x, pages_after[x])
```

```python
def is_safe(atualization, pages_before):
    for i, page in enumerate(atualization):
        for j, other_page in enumerate(atualization):
            if page in pages_before[other_page] and i > j:
                return False
    return True


accumulated = 0
for atualization in atualizations:
    atualization = list(map(int, atualization.split(',')))
    safe = is_safe(atualization, pages_before)
    if safe:
        mid = len(atualization) // 2
        accumulated += atualization[mid] # se for safe, adicionar o valor do meio
print("Part 1", accumulated) # part 1


def insert_page(page, ordered_sequence, pages_before, pages_after):
    if page in ordered_sequence:
        return ordered_sequence
    # if not ordered_sequence:
    #     return [page]
    for i, other_page in enumerate(ordered_sequence):
        if other_page in pages_before[page]:
            return ordered_sequence[:i] + [page] + ordered_sequence[i:]
    return ordered_sequence + [page]

def make_sequence_valid(sequence, pages_before, pages_after):
    ordered_sequence = []
    for page in sequence:
        ordered_sequence = insert_page(page, ordered_sequence, pages_before, pages_after)
    return ordered_sequence

accumulated = 0
for atualization in atualizations:
    atualization = list(map(int, atualization.split(',')))
    safe = is_safe(atualization, pages_before)
    if not safe:
        mid = len(atualization) // 2
        ordered_sequence = make_sequence_valid(atualization, pages_before, pages_after)
        accumulated += ordered_sequence[mid] # se for safe, adicionar o valor do meio
print("Part 2", accumulated)

```
