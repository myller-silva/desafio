```python
def find_robot(grid):
    for row_index, row in enumerate(grid):
        for col_index, cell in enumerate(row):
            if cell == "@":
                return [row_index, col_index]


def move_robot(grid, robot_pos, move):
    row, col = robot_pos
    directions = {"^": (-1, 0), "v": (1, 0), "<": (0, -1), ">": (0, 1)} # row, col
    direction = directions[move]
    new_row, new_col = row, col
    new_row += direction[0]
    new_col += direction[1]
    
    if grid[new_row][new_col] == ".":
        grid[new_row][new_col] = "@"
        grid[row][col] = "."
        robot_pos = (new_row, new_col)
    
    if grid[new_row][new_col] == "O": # mover a caixa para a proxima posição que não é caixa ou parede
        temp_row, temp_col = new_row, new_col
        while grid[temp_row][temp_col] == "O":
            temp_row += direction[0]
            temp_col += direction[1]
            if grid[temp_row][temp_col] == "#":
                break
            if grid[temp_row][temp_col] == "O":
                continue
            if grid[temp_row][temp_col] == ".":
                grid[temp_row][temp_col] = "O"
                grid[new_row][new_col] = "@"
                grid[row][col] = "."
                robot_pos = (new_row, new_col)
                break
    return robot_pos



def calculate_gps_sum(grid):
    gps_sum = 0
    for row_index, row in enumerate(grid):
        for col_index, cell in enumerate(row):
            if cell == "O":
                gps_sum += (row_index * 100) + col_index
    return gps_sum

```

```python
with open("inputs/15.txt", "r") as file:
    lines = file.readlines()
grid = [list(line.strip()) for line in lines if line.startswith("#")]
moves = "".join(line.strip() for line in lines if not line.startswith("#"))
robot_pos = find_robot(grid)
for move in moves:
    robot_pos = move_robot(grid, robot_pos, move)

gps_sum = calculate_gps_sum(grid)
print(f"A soma das coordenadas GPS de todas as caixas é: {gps_sum}")
```

```python
def transform_grid(grid):
    new_grid = []
    for i in range(len(grid)):
        modified_row = ""
        for j in range(len(grid[i])):
            symbol_mapping = {"#": "##", "O": "[]", ".": "..", "@": "@."}
            modified_row += symbol_mapping[grid[i][j]]
        new_grid.append(list(modified_row))
    return new_grid

with open("inputs/15.txt", "r") as file:
    lines = file.readlines()
grid = [list(line.strip()) for line in lines if line.startswith("#")]

new_grid = transform_grid(grid)

for row in new_grid:
    # print(len(row))
    print("".join(row))

```
