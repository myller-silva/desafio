```python
from collections import deque

def count_paths(matrix):
    rows = len(matrix)
    # cols = max(len(row) for row in matrix)  # para garantir retangularidade
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    results = {}

    def bfs(start):
        queue = deque([(start[0], start[1], 0)])
        visited = set()
        visited.add((start[0], start[1]))

        paths_count = 0
        while queue:
            x, y, value = queue.popleft()
            
            if int(matrix[x][y]) == 9:
                paths_count += 1
                continue
            
            for dx, dy in directions:
                nx, ny = x + dx, y + dy
                # verificacao rigorosa de limites
                if 0 <= nx < rows and 0 <= ny < len(matrix[nx]) and (nx, ny) not in visited:
                    if matrix[nx][ny].isdigit() and int(matrix[nx][ny]) == value + 1:
                        visited.add((nx, ny))
                        queue.append((nx, ny, value + 1))
        
        return paths_count

    for r in range(rows):
        for c in range(len(matrix[r])):
            if matrix[r][c].isdigit() and int(matrix[r][c]) == 0:
                results[(r, c)] = bfs((r, c))
    
    return results


matrix = content.strip().split("\n")
matrix = [list(row) for row in matrix]

result = count_paths(matrix)
print(result)

import numpy as np
np.sum(list(result.values()))

```

```python
from collections import deque

def count_paths(matrix):
    rows = len(matrix)
    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
    results = {}

    def bfs(start):
        queue = deque([(start[0], start[1], 0)])
        paths_count = 0
        
        while queue:
            x, y, value = queue.popleft()

            if int(matrix[x][y]) == 9:
                paths_count += 1
                continue
            
            for dx, dy in directions:
                nx, ny = x + dx, y + dy
                if 0 <= nx < rows and 0 <= ny < len(matrix[nx]):
                    if matrix[nx][ny].isdigit() and int(matrix[nx][ny]) == value + 1:
                        queue.append((nx, ny, value + 1))
        
        return paths_count

    # identifica os trailheads e calcula a contagem de trilhas para cada um
    for r in range(rows):
        for c in range(len(matrix[r])):
            if matrix[r][c].isdigit() and int(matrix[r][c]) == 0:
                results[(r, c)] = bfs((r, c))
    
    return results


matrix = content.strip().split("\n")
matrix = [list(row) for row in matrix]

result = count_paths(matrix)
print(result)

import numpy as np
print(np.sum(list(result.values())))
```
