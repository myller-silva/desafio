```python
def read_file(filename: str) -> list:
    result = []
    with open(filename) as f:
        for line in f:
            temp = []
            for elem in line.strip():
                temp.append(elem)
            result.append(temp)
    return result
content = read_file('inputs/8.txt')
```

```python
for row in content:
    print(''.join(row))
```

```python
R = len(content)
C = len(content[0])

def get_antenna_positions(content):
    antennas = dict()
    for i in range(R):
        for j in range(C):
            antenna = content[i][j]
            pos = (i, j)
            if antenna.isalpha() or antenna.isdigit():
                if antenna in antennas:
                    antennas[antenna].append(pos)
                else: 
                    antennas[antenna] = [pos]
    return antennas

antennas = get_antenna_positions(content)

antennas
```

Para cada par de antena da mesma frequencia, calcular onde ficaria os dois antinode

```python
def get_antinodes(p1, p2):
    if p1[0] > p2[0]:
        p1, p2 = p2, p1
        
    (x1, y1), (x2, y2) = p1, p2
    
    dx, dy = x2 - x1, y2 - y1
    
    return (x1 - dx, y1 - dy), (x2 + dx, y2 + dy)

def get_all_antinodes(p1, p2, grid_size):
    
    if p1[0] > p2[0]: # se a primeira antena estiver mais a direita, trocar
        p1, p2 = p2, p1
    
    (x1, y1), (x2, y2) = p1, p2
    
    dx, dy = x2 - x1, y2 - y1
    
    nodes = set()
    
    # pegar os pontos antes de p1
    p = (x1, y1)
    while p[0] >= 0 and p[1] >= 0 and p[0] < grid_size[0] and p[1] < grid_size[1]:
        nodes.add(p)
        p = (p[0] - dx, p[1] - dy)
    
    # pegar os pontos depois de p2
    p = (x2, y2)
    while p[0] >= 0 and p[1] >= 0 and p[0] < grid_size[0] and p[1] < grid_size[1]:
        nodes.add(p)
        p = (p[0] + dx, p[1] + dy)
    
    return nodes

part1 = set()
part2 = set()

for (k, positions) in antennas.items():
    for (x1, y1) in positions: 
        for (x2, y2) in positions: 
            if (x1, y1) != (x2, y2): # se nao for a mesma antena
                
                # parte 1
                q1, q2 = get_antinodes((x1, y1), (x2, y2))
                
                if (0 <= q1[0] < R and 0 <= q1[1] < C):
                    part1.add(q1)  
                
                if (0 <= q2[0] < R and 0 <= q2[1] < C):
                    part1.add(q2)
                
                # parte 2 
                antinodes = get_all_antinodes((x1, y1), (x2, y2), (R, C))
                part2.update(antinodes)

print("Part 1:", len(part1))
print("Part 2:", len(part2))
```
