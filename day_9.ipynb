```python
# content = list("2333133121414131402")
def get_next(content):
    c = 0
    # new_content = ''
    new_content = list()
    for i in range(len(content)):
        char = content[i]
        if i % 2 == 0:  # Índice par: caractere repetido
            # new_content.append(f'{c}' * int(char))
            new_content.extend(f'{c}' for _ in range(int(char)))
            c += 1
        else:
            new_content.extend('.' for _ in range(int(char)) )
    return new_content

next_content = get_next(content)
# next_content

for e in next_content:
    print(e, end='')

```

```python
# next_content = list(next_content)

for i in range(len(next_content)):
    element1 = next_content[i]
    if element1 != '.':
        continue
    
    # percorrer a lista decrescente até i
    for j in range(len(next_content)-1, i, -1):
        element2 = next_content[j]
        if element2 == '.':
            continue
        next_content[i], next_content[j] = next_content[j], next_content[i]
        break
    # print(''.join(next_content))
```

```python
checksum = 0

for i in range(len(next_content)):
    if '.' in next_content[i]:
        continue
    if next_content[i] == '':
        continue
    checksum += int(next_content[i]) * i
checksum
```

```python
content = read_file('inputs/9.txt')
content = content[:-1]
next_content = get_next(content)

for i in range(len(next_content)):
    print(next_content[i], end='')
```

```python

# next_content = list("00...111...2...333.44.5555.6666.777.888899")


def get_last_empty_space(array, length, init, finish):
    tam = 0
    for i in range(0, init):
        if array[i] == '.':
            tam += 1
        if tam == length:
            return i
        if array[i] != '.':
            tam = 0
    return None


init = None
finish = None


for i in range(len(next_content)-1, -1, -1):
    # print(i, end=' ')
    # pegar o ultimo elemento
    if next_content[i] == '.':
        continue
    
    if finish is None:
        finish = i
    if next_content[i] == next_content[i-1]:
        continue
    
    init = i
    
    length = finish - init + 1
    # print(length, init, finish, t[init:finish+1])
    empty_space = get_last_empty_space(next_content, length, init, finish)
    if empty_space is not None:
        # preencher o espaço vazio com o elemento
        for j in range(init, finish+1):
            next_content[empty_space] = next_content[j]
            next_content[j] = '.'
            empty_space -= 1
    
    # print(''.join(next_content))
    init = None
    finish = None

```

```python
checksum = 0

for i in range(len(next_content)):
    if '.' in next_content[i]:
        continue
    if next_content[i] == '':
        continue
    checksum += int(next_content[i]) * i
checksum
```
