```python
def read_file(file):
    left, right = [], []
    with open(file, 'r') as f:
        for line in f:
            x, y = line.split()
            left.append(int(x))
            right.append(int(y))
    return left, right
```

```python
left, right = read_file('inputs/1.txt')
```

# Part 1

```python
left, right = sorted(left), sorted(right)
content = list(zip(left, right))
content = list(content)
content
```

```python
distance = lambda x, y: abs(x - y)

#somatorio de todas as distancias
somatorio = sum([distance(x, y) for x, y in content])
somatorio

```

# Part 2

```python
# contar a quantidade de vezes que cada numero do left aparece em right

counts = {}
for x in left:
    counts[x] = 0
    for y in right:
        if x == y:
            counts[x] += 1

counts
# fazer multiplicacao de chave pelo valor
soma = sum([x * y for x, y in counts.items()])
soma
```
