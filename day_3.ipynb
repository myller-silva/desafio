# Part 1

```python
def ler_arquivo(file_name):
    content = ""
    with open(file_name, 'r') as file:
        for line in file:
            content += line
    return content
```

```python
file_name = 'inputs/3.txt'
content = ler_arquivo(file_name)
```

```python
import re 

def parse_input(content, pattern):
    return re.findall(pattern, content)


pattern = r'mul\((\d{1,3}),(\d{1,3})\)'
claims = re.findall(pattern, content)
claims
```

```python
def calculate_multiplication(claims):
    result = 0
    for claim in claims:
        result += int(claim[0]) * int(claim[1])
    return result

calculate_multiplication(claims)
```

# Part 2

```python
def remove(string):
    result = ""
    init = string.find("don't()")
    while init != -1:
        end = string.find("do()", init)
        if end == -1:
            break
        result += string[:init]
        string = string[end + 3:]  # Adiciona 3 para pular "do()"
        init = string.find("don't()")
    result += string
    return result


# Teste
teste = "xmul(2,4)&mul[3,7]!^dodon't()_mul(5,5)+mul(32,64](mul(11,8)undo()?mul(8,5))don't()mul(8,5)do()mul(8,5)"
content_teste = remove(teste)
print(content_teste)
```

```python
content_2 = remove(content)

pattern = r'mul\((\d{1,3}),(\d{1,3})\)'
claims =  re.findall(pattern, content_2)
calculate_multiplication(claims)
```

# Part 2 with Regex

```python
import re


# Exemplo de uso
start_str = "don't()"
end_str = "do()"
regex_pattern = f"{re.escape(start_str)}.*?{re.escape(end_str)}"
print(regex_pattern)

new_content = re.sub(regex_pattern, "", content)

claims = re.findall(pattern, new_content)
calculate_multiplication(claims)
```
